Набор данных представляет собой статистику параметров автомобилей на вторичном рынке в Молдавии. 

Набор включает ряд категориальных и численных значений, составляющих одну запись (строку). Число записей можно найти как число строк.

# Поиск Дубликатов
 


методы .duplicated() и .sum()

```python
df.duplicated().sum() # вычисляем сколько всего дубликатов 
```

**Удаление дубликатов**

Методы .drop_duplicates() и .reset_index()

```python
DF = df.drop_duplicates()# убиравем дубликаты 

DF = DF.reset_index(drop=True)# делаем список обновляя индексацию с убранными дубликатами  и выводим его 

DF
```

```python
DF.to_csv('/content/cars_moldova_no_dup.csv',index=False) #. сохраняем все в файл 
```
